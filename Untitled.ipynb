```python
df = pandas.read_csv('/Users/tarsisazevedo/Downloads/Avaliação da Disciplina de Engenharia de Dados (respostas) - Respostas ao formulário 1.csv')
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
ax = plot_likert.plot_likert(
    df["Os assuntos tratados são relevantes no dia a dia da indústria e sociedade"], 
    range(1,6), 
    plot_percentage=True, 
    figsize=(14, 4)
)
for bars, color in zip(ax.containers[1:], ['white'] + ['black'] * 2 + ['white'] * 2):
    ax.bar_label(bars, label_type='center', fmt='%.1f %%', color=color, fontsize=15)

plt.tight_layout()
plt.show()
```
